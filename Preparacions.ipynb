```python
dades = pd.merge(dades, comarques, on='clau_2', how='left')
```

```python
dades.drop(labels=['Comarca_y','Any_y','Codi','Literal','clau_2','Població_y'], axis=1 , inplace=True)
dades.rename(columns = {'Any_x':'Any',
                        'Homes. De 0 a 14 anys':'H0-14',
                        'Homes. De 15 a 64 anys':'H15-64',
                        'Homes. De 65 anys i més':'H65-X',
                        'Dones. De 0 a 14 anys':'D0-14',
                        'Dones. De 15 a 64 anys':'D15-64',
                        'Dones. De 65 anys i més':'D65-X',
                        'Total. De 0 a 14 anys':'T0-14',
                        'Total. De 15 a 64 anys':'T15-64',
                        'Total. De 65 anys i més':'T65-X',
                        'Població_x':'Població municipi',
                        'Comarca_x':'Comarca',
                        'Nombre de municipis':'municipis/comarca',
                        'Capital de comarca':'Capital comarca',
                        'Superfície(km2)':'Superfície municipi(km2)'}, inplace = True)
```

```python
col = ['Any','Municipi', 'Altitud(m)', 'Superfície municipi(km2)', 'Comarca', 'tipus', 'Capital comarca', 
       'municipis/comarca', 'Província',
       'H0-14', 'H15-64', 'H65-X', 'D0-14', 'D15-64', 'D65-X', 'T0-14', 'T15-64', 'T65-X', 
       'Població municipi']
dades = dades.loc[:, col]
```

```python
dades.loc[ dades['Província'] == 'Girona/Barcelona (4)', 'Província'] = 'Girona'
dades.loc[ dades['Província'] == 'Barcelona/Girona (3)', 'Província'] = 'Barcelona'
dades.loc[ dades['Província'] == 'Barcelona/Lleida (1)', 'Província'] = 'Barcelona'
dades.loc[ dades['Província'] == 'Girona/Lleida (2)', 'Província'] = 'Girona'
dades.loc[ dades['Província'] == 'Taragona', 'Província'] = 'Tarragona'

dades.loc[ dades['Comarca'] == "Val d'Aran", 'Capital comarca'] = 'Vielha'
dades.loc[ dades['Comarca'] == "Val d'Aran", 'municipis/comarca'] = 9
dades.loc[ dades['Comarca'] == "Val d'Aran", 'Província'] = 'Lleida'
dades.loc[ dades['Comarca'] == "Val d'Aran", 'tipus'] = 'Pirineu'
```

```python
dades['municipis/comarca'] = dades['municipis/comarca'].astype(int)
dades['Altitud(m)'] = dades['Altitud(m)'].str.replace(" ","").astype(int)
dades['Superfície municipi(km2)'] = dades['Superfície municipi(km2)'].str.replace(',','.', regex=True).astype(float)
dades['Població municipi'] = dades['Població municipi'].str.replace('.','', regex=True).astype(int)
dades['H0-14'] = dades['H0-14'].str.replace(".","", regex=True).astype(int)
dades['H15-64'] = dades['H15-64'].str.replace(".","", regex=True).astype(int)
dades['H65-X'] = dades['H65-X'].str.replace(".","", regex=True).astype(int)
dades['D0-14'] = dades['D0-14'].str.replace(".","", regex=True).astype(int)
dades['D15-64'] = dades['D15-64'].str.replace(".","", regex=True).astype(int)
dades['D65-X'] = dades['D65-X'].str.replace(".","", regex=True).astype(int)
dades['T0-14'] = dades['T0-14'].str.replace(".","", regex=True).astype(int)
dades['T15-64'] = dades['T15-64'].str.replace(".","", regex=True).astype(int)
dades['T65-X'] = dades['T65-X'].str.replace(".","", regex=True).astype(int)
```

```python
#dades.dtypes
```

```python
dades['THomes'] = dades['H0-14'] + dades['H15-64'] + dades['H65-X']
dades['TDones'] = dades['D0-14'] + dades['D15-64'] + dades['D65-X']
dades['HShare'] = (dades['THomes'] / dades['Població municipi']).round(decimals=4)
dades['DShare'] = (dades['TDones'] / dades['Població municipi']).round(decimals=4)
dades['T0-14Share'] = (dades['T0-14'] / (dades['T0-14'] + dades['T15-64'] + dades['T65-X'])).round(decimals=4)
dades['T15-64Share'] = (dades['T15-64'] / (dades['T0-14'] + dades['T15-64'] + dades['T65-X'])).round(decimals=4)
dades['T65-XShare'] = (dades['T65-X'] / (dades['T0-14'] + dades['T15-64'] + dades['T65-X'])).round(decimals=4)
dades['Densitat (població/km2)'] = (dades['Població municipi'] / dades['Superfície municipi(km2)']).round(decimals=2)
```
